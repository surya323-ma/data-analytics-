*Visualize and analyse the maxi,mini and medium income*

```python
y=list(df.income)
plt.boxplot(y)
plt.show()
```

```python

numeric_df = df.select_dtypes(include=np.number)
numeric_df.groupby([df['gender'], df['reduced']]).mean()
```

*Visualize is there is any missing values in the dataset based ona heat map*

```python
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
```

```python
plt.figure(figsize=(10,10))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cbar=True,cmap='Blues')
plt.show()
```

Analyse how the income of a patient affect the num of visits to the *hospital*

```python
plt.figure(figsize=(10,10))
plt.scatter(x='income',y='visits',data=df)
plt.xlabel('income')
plt.ylabel('visits')
```

```python
sns.histplot(df.gender,bins=4)
```

```python
label='yes','no'
y_df=df[df['freepoor']=='yes']
n_df=df[df['freepoor']=='no']
x=[y_df.shape[0],n_df.shape[0]]
plt.figure(figsize=(5,10))
plt.pie(x,labels=label,autopct='%1.2f%%')
plt.title("% of people getting govt health insurance due to low income")
plt.show()
```

```python

label='yes','no'
y_df=df[df['private']=='yes']
n_df=df[df['private']=='no']
x=[y_df.shape[0],n_df.shape[0]]
plt.figure(figsize=(5,10))
plt.pie(x,labels=label,autopct='%1.2f%%')
plt.title("% of people having private health insurance ")
plt.show()
```

```python
label='yes','no'
y_df=df[df['freerepat']=='yes']
n_df=df[df['freerepat']=='no']
x=[y_df.shape[0],n_df.shape[0]]
plt.figure(figsize=(5,10))
plt.pie(x,labels=label,autopct='%0.2f%%')
plt.title("% of people getting govt health insurance due to low age,didability or veteran status")
plt.show()
```

```python
db=df.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.barh(db['gender'],db['reduced'],color=['cornflowerblue','lightseagreen'])
plt.title('bar chart')
plt.xlabel('gender')
plt.ylabel('reduced activity')
plt.show()
```
