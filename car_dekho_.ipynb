## Analyze 'year', 'selling price', missing values, and total records



```python
min_year = df['Year'].min()
max_year = df['Year'].max()
lowest_selling_price = df['Selling_Price'].min()
highest_selling_price = df['Selling_Price'].max()
num_records = len(df)
missing_values = df.isnull().sum().sum()

print(f"1. Vehicles are present from manufacturing year {min_year} to {max_year}.")
print(f"2. The lowest price a vehicle is sold for is {lowest_selling_price}.")
print(f"3. The highest price a vehicle is sold for is {highest_selling_price}.")
print(f"4. There are {num_records} records in this data.")
print(f"5. There are {missing_values} missing records in this data.")
```

## Analyze 'car name' and 'fuel type'

### Subtask:
Answer questions 6-8 by analyzing the 'Car_Name' and 'Fuel_Type' columns to find the number of unique vehicles, the most sold vehicle, and the count of CNG vehicles.


```python
num_unique_cars = df['Car_Name'].nunique()
most_sold_car = df['Car_Name'].mode()[0]
cng_vehicles_count = df[df['Fuel_Type'] == 'CNG'].shape[0]

print(f"6. There are {num_unique_cars} different vehicles present in this data.")
print(f"7. The most sold vehicle in this data is {most_sold_car}.")
print(f"8. Yes, the database includes {cng_vehicles_count} CNG vehicles.")
```

## Analyze 'seller type', 'transmission', and 'owner'

### Subtask:
Answer questions 9-11 by analyzing the 'Seller_Type', 'Transmission', and 'Owner' columns to find the number of vehicles from individuals, with auto transmission, and single-person ownership.


```python
individual_sellers_count = df[df['Seller_Type'] == 'Individual'].shape[0]
auto_transmission_count = df[df['Transmission'] == 'Automatic'].shape[0]
single_owner_count = df[df['Owner'] == 0].shape[0]

print(f"9. There are {individual_sellers_count} vehicles here for sell from Individuals directly.")
print(f"10. Yes, this database contains {auto_transmission_count} auto transmission vehicles.")
print(f"11. There are {single_owner_count} single person owned vehicles in this database.")
```

## Calculate cost depreciation and identify depreciation trends

### Subtask:
Answer questions 12-13 by calculating cost depreciation and identifying the most/least depreciated vehicles and brands less affected by depreciation.


**Reasoning**:
Calculate the depreciation, find the most and least depreciated cars, group by car name to calculate average depreciation, and sort to find brands less affected by depreciation.



```python
df['Depreciation'] = df['Present_Price'] - df['Selling_Price']

most_depreciated_car = df.loc[df['Depreciation'].idxmax()]
least_depreciated_car = df.loc[df['Depreciation'].idxmin()]

print(f"12. The most depreciated vehicle is: {most_depreciated_car['Car_Name']} with depreciation of {most_depreciated_car['Depreciation']:.2f}")
print(f"The least depreciated vehicle is: {least_depreciated_car['Car_Name']} with depreciation of {least_depreciated_car['Depreciation']:.2f}")

average_depreciation_by_brand = df.groupby('Car_Name')['Depreciation'].mean().sort_values()

print("\n13. Brands of vehicles less affected by cost depreciation (sorted by average depreciation):")
display(average_depreciation_by_brand)
```

## Analyze factors affecting cost depreciation

### Subtask:
Answer question 14 by analyzing the factors that might affect cost depreciation based on the data.


**Reasoning**:
Examine the relationships between 'Depreciation' and other numerical and categorical columns through visualizations to identify potential factors affecting depreciation.



```python
# Analyze correlation with numerical columns
correlation_matrix = df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven', 'Depreciation']].corr()
print("Correlation matrix of numerical columns and Depreciation:")
display(correlation_matrix['Depreciation'].sort_values(ascending=False))

# Visualize relationship between numerical features and Depreciation
plt.figure(figsize=(15, 5))
plt.subplot(1, 4, 1)
sns.scatterplot(x='Year', y='Depreciation', data=df)
plt.title('Depreciation vs Year')

plt.subplot(1, 4, 2)
sns.scatterplot(x='Selling_Price', y='Depreciation', data=df)
plt.title('Depreciation vs Selling_Price')

plt.subplot(1, 4, 3)
sns.scatterplot(x='Present_Price', y='Depreciation', data=df)
plt.title('Depreciation vs Present_Price')

plt.subplot(1, 4, 4)
sns.scatterplot(x='Kms_Driven', y='Depreciation', data=df)
plt.title('Depreciation vs Kms_Driven')
plt.tight_layout()
plt.show()

# Visualize relationship between categorical features and Depreciation
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.boxplot(x='Fuel_Type', y='Depreciation', data=df)
plt.title('Depreciation by Fuel Type')

plt.subplot(1, 3, 2)
sns.boxplot(x='Seller_Type', y='Depreciation', data=df)
plt.title('Depreciation by Seller Type')

plt.subplot(1, 3, 3)
sns.boxplot(x='Transmission', y='Depreciation', data=df)
plt.title('Depreciation by Transmission')
plt.tight_layout()
plt.show()
```
