```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
from google.colab import files
uploaded=files.upload()
file_name = next(iter(uploaded))
df=pd.read_excel(file_name)
print(df.columns)
df.head()
```

```python
df.info()
```

```python
df.duplicated().value_counts()
```

```python
df[df['neighbourhood group']=='brookln']
```

```python
df.drop_duplicates(inplace=True)
df.drop(['house_rules','license'],axis=1,inplace=True)
```

```python
df['price'] = df['price'].astype(str).str.replace('$','',regex=False)
df['service fee'] = df['service fee'].astype(str).str.replace('$','',regex=False)
```

```python
df.rename(columns={'service fee':'service_fee_$','price':'price_$'},inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['price_$']=df['price_$'].astype(float)
df['service_fee_$']=df['service_fee_$'].astype(float)
df['id']=df['id'].astype(str)
```

```python
df['host id']=df['host id'].astype(str)
df['last review']=pd.to_datetime(df['last review'])
df['Construction year']=df['Construction year'].astype(int)
```

```python
df.loc[df['neighbourhood group'] == 'brookln', 'neighbourhood group'] = 'Brooklyn'
df = df.drop(df[df['availability 365'] > 500].index)
```

EXPLORATORY DATA ANALYSIS(EDA)

```python
df.describe()
```

```python
property_type = df['room type'].value_counts().to_frame()
property_type
```

```python
room_type_bar=plt.bar(property_type.index,property_type.loc[:,"count"]);
plt.bar_label(room_type_bar,labels=property_type.loc[:,"count"],padding=4);
plt.ylim([0,50000]);
plt.xlabel('Room Type');
plt.ylabel('Room Type Count');
plt.title('Property types and their count in the Dataset');
```

```python
hood_group = df['neighbourhood group'].value_counts().to_frame()
hood_group
```

```python
hood_group_bar=plt.bar(hood_group.index,hood_group.loc[:,"count"]);
plt.bar_label(hood_group_bar,labels=hood_group.loc[:,"count"],padding=4);
plt.ylim([0,50000]);
plt.xlabel('Neighbourhood Group');
plt.ylabel('Number of listings');
plt.xticks(rotation=45);
plt.title('Number of listings in each Neighbourhood Group');
```

```python
avg_price = df.groupby('neighbourhood group')['price_$'].mean().sort_values(ascending=False).to_frame()
avg_price_bar=plt.bar(avg_price.index,avg_price.loc[:,"price_$"]);
plt.bar_label(avg_price_bar,labels=round(avg_price.loc[:,"price_$"],4),padding=4,label_type='edge');
plt.ylim([0,700]);
plt.xlabel('Neighbourhood Group');
plt.ylabel('Average Price per listing($)');
plt.title('Average Price per listing in each Neighbourhood Group')
plt.xticks(rotation=45);
```

```python
df.groupby(df['Construction year'])['price_$'].mean().to_frame().plot();
plt.xlabel('Construction Year');
plt.ylabel('Average Price per listing($)');
plt.title('Average Price per listing in each Construction Year');
```

```python
hosts=df.groupby('host id')['calculated host listings count'].sum().sort_values(ascending=False).nlargest(10).to_frame()
hosts_bar=plt.bar(hosts.index,hosts.loc[:,"calculated host listings count"]);
plt.bar_label(hosts_bar,labels=hosts.loc[:,"calculated host listings count"],padding=9,label_type='edge');
plt.ylim([0,120000]);
plt.xlabel('Host name');
plt.ylabel('Cal.host Listings count');
plt.xticks(rotation=75);
plt.title('Top 10 Hosts by cal. host  listings count');
```

```python
reviews=df.groupby('host_identity_verified')['review rate number'].mean().sort_values(ascending=False).to_frame()
reviews
```

```python
review_bar=plt.bar(reviews.index,reviews.loc[:,"review rate number"]);
plt.bar_label(review_bar,labels=round(reviews.loc[:,"review rate number"],3),padding=6);
plt.ylim([0,6]);
plt.xlabel('Host  Verified status');
plt.ylabel('Average Review Rate number');
plt.title('Average Review Rate for each Verifications statistics');
```

```python
sns.boxplot(data=df,x='host_identity_verified',y='review rate number',color='skyblue');
plt.xlabel('host verfication status');
plt.ylabel('Average Review Rate number');
plt.title('Average Review Rate for each Verifications statistics');
```

```python
df['price_$'].corr(df['service_fee_$'])
```
